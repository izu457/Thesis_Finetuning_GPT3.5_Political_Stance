Create Training Files

```python
# set seed
np.random.seed(1)

# Add EP9 to Vote ID to avoid duplicated Vote IDs
voted_docs_final_EP9.index = voted_docs_final_EP9.index.astype(str) + 'EP9'
voted_docs_final_EP8.index = voted_docs_final_EP8.index.astype(str) + 'EP8'
voted_docs_final_EP7.index = voted_docs_final_EP7.index.astype(str) + 'EP7'

# split the data into training and testing
training_EP9, testing_EP9 = fu.split_select_data(voted_docs_final_EP9, 0.8)
training_EP8, testing_EP8 = fu.split_select_data(voted_docs_final_EP8, 0.8)
training_EP7, testing_EP7 = fu.split_select_data(voted_docs_final_EP7, 0.8)
display(training_EP9.head())

# split the training data into training and baseline
pure_training_EP9, baseline_EP9 = fu.split_select_data(training_EP9, 0.95)
pure_training_EP8, baseline_EP8 = fu.split_select_data(training_EP8, 0.95)
pure_training_EP7, baseline_EP7 = fu.split_select_data(training_EP7, 0.95)

# split the testing data into testing and validation
testing_EP9, validation_EP9 = fu.split_select_data(testing_EP9, 0.5)
testing_EP8, validation_EP8 = fu.split_select_data(testing_EP8, 0.5) 
testing_EP7, validation_EP7 = fu.split_select_data(testing_EP7, 0.5)

# join training and testing data
training = pd.concat([training_EP9, training_EP8, training_EP7])
testing = pd.concat([testing_EP9, testing_EP8, testing_EP7])
validation = pd.concat([validation_EP9, validation_EP8, validation_EP7])
# make column 2 to 13 floats
training.iloc[:, 1:9] = training.iloc[:, 1:9].astype(float)	
testing.iloc[:, 1:9] = testing.iloc[:, 1:9].astype(float)
validation.iloc[:, 1:9] = validation.iloc[:, 1:9].astype(float)
# make sure data is shuffled
training = training.sample(frac=1)
testing = testing.sample(frac=1)
validation = validation.sample(frac=1)
display(training.head())
# save to csv
training.to_csv('data/finetuning_data/training_data.csv')
testing.to_csv('data/finetuning_data/testing_data.csv')
validation.to_csv('data/finetuning_data/validation_data.csv')
```
